```python
df = pd.read_excel (r'/Users/ayaoji/Desktop/Summer Assignment 2021/japan-data.xlsx', parse_dates=["Year"])
df
```

## 1) Last Observation Carried Forwards (forward fill)

```python
LOCF_df = df.fillna(method='ffill')
LOCF_df
```

## 2) Next Observation Carried Backwards 

```python
BOCB_df = df.fillna(method='backfill')
BOCB_df
```

## 3) Liner Interpolation 

```python
interp_df = df.interpolate()
interp_df
```

## 4) Spline

```python
spline_df = df.interpolate(methos='spline')
spline_df
```
